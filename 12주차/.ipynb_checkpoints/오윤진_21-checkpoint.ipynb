## 21.3. 임포트 타임과 런타임

## 21.3.1. 코드 평가 시점 연습문제

시나리오 #1 : 파이썬 콘솔에서 evaltime.py 모듈 임포트
  import evaltime 문으로 실행
  인터프리터는 임포트된 모듈과 이 모듈이 임포트하는 모듈에 들어있는 모든 클래스를 실행
  장식된 클래스에 붙어 있는 데커레이터 함수를 실행하기 전에 장식된 클래스의 본체를 먼저 평가. 데커레이터가 처리할 클래스 객체를 받아야 하므로 클래스 객체 먼저 생성해야.
  이 시나리오에서 사용자 정의 함수/메서드 중 deco_alpha() 데커레이터만 유일하게 실행됨.
  
시나리오 #2 : 명령행에서 evaltime.py 모듈 실행
  클래스 데커레이터가 변경한 내용이 서브클래스에 영향을 주지 않을 수도 잇음.

## 21.4. 메타클래스 기본 지식

메타클래스 : 클래스 팩토리. but 함수 대신 클래스로 만들어짐.
             공장이 또 다른 공장을 만든다. (클래스 -> 클래스 생성)

클래스는 다른 어떤 클래스의 객체여야.
파이썬 클래스 : type의 객체
type : 대부분의 내장된 클래스, 사용자 정의 클래스에 대한 메타 클래스임.

무한 회귀 방지 : type은 자기 자신의 객체로 정의

메타클래스 : type, ABCMeta, Enum

collections.Iterable : abc.ABCMeta
Iterable : 추상클래스, ABCMeta 아님. ABCMeta 클래스의 객체

ABCMeta 클래스 : type

모든 클래스 : 직간접적으로 type의 객체
메타클래스 : type의 서브클래스 >> 클래스 팩토리로 행동. 특히 __init__() 구현해 자신의 객체 커스터마이즈 할 수 있음.

## 21.6. 메타클래스 __prepare__() 특별 메서드

클래스 속성 정의되는 순서 알아야하는 경우 있음

type() 생성자 & 메타클래스의 __new__(), __init__() : 이름,속성의 매핑으로 평가된 클래스 본체 받음.
    매핑: 딕셔너리형 >> 메타클래스/데커레이터가 확인 가능 >> 클래스 본체 안에 등장하는 속성의 순서 사라짐.
    
__prepare__() : 특별 메서드. 메타클래스에서만 의미 있음. 클래스 메서드이어야 함.(*@classmethod 데커레이터로 장식해야*)

인터프리터 : 메타클래스 __new__() 호출 전에 __prepare__() 호출

__prepare__() : 메타클래스를 첫번째 인수로 전달받음. 생성할 클래스명, 슈퍼클래스 담긴 튜플 받음. __new__(), __init__()의 마지막 인수로 전달할 매핑을 반환해야.


실제 메타클래스의 이용
  속성 검증
  많은 메서드에 데커레이터를 일괄 적용
  객체 ㄱ직렬화 및 데이터 변환
  객체 관계 매핑
  객체 기반 영속성
  다른 언어에서 만든 클래스 구조체를  파이썬 클래스로 동적 변환
 
## 21.7. 객체로서의 클래스

cls.__bases__ : 슈퍼클래스들을 담은 튜플
cls.__qualname__ : 전역 범위에서 클래스 정의를 담은 모듈까지의 경로를 점으로 구분한 크ㅡㄹ래스/함수의 경로명을 담고 있는 속성.
cls.__subclasses__() : 현재 메모리에 존재하는 클래스의 바로 아래 서브클래스들의 리스트 반환.
cls.mro() : 인터프리터는 이 메서드를 클래스의 __mro__ 속성에 담겨 있는 슈퍼클래스의 튜플을 가져와 클래스를 생성할 때 호출함.
