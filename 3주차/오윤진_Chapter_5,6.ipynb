### 5.2.1. map(), filter(), reduce()의 대안

지능형 리스트나 제너레이트 표현식으로 대체 가능

파이썬 3에서 map(), filter() 제너레이터 반환 >> 제너레이터 표현식으로 대체 가능
reduce() 합계 구할 때 사용 >> sum()
*sum(), reduce()는 이전 결과 누적시켜 일련의 값을 하나의 값으로 리덕션

리덕션 함수 : all(iterable), any(iterable)
all(iterable) : 모든 iterable이 참이면 True 반환, all([])>>True
any(iterable) : iterable 중 하나라도 참이면 True 반환, any([])>>False

## 5.4. 일곱가지 콜러블 ( callable() )

callable() : 호출할 수 있는 객체인지 판단하는 내장 함수

7가지 콜러블 
1. 사용자 정의 함수 : def 문 / 람다 표현식으로 생성
2. 내장 함수 : c언어로 구현된 함수 ex) len(), time.strftime()
3. 내장 메서드 : c언어로 구현된 메서드 ex) dict.get()
4. 메서드 : 클래스 본체에 정의된 함수
5. 클래스 : 클래스 호출 >> 자신의 __new__() 메서드 실행해 객체 생성 >> __init__()로 초기화 >> 호출자에 객체 반환.
            파이썬에 new 연산자 없음 ; 클래스 호출== 함수 호출
6. 클래스 객체 : 클래스가 __call__() 메서드 구현 >> 클래스 객체 함수로 호출 가능
7. 제너레이터 함수 : yield 키워드 사용하는 함수나  메서드. 호출 >> 제너레이터 객체 반환

## 5.6. 함수 인트로스펙션

함수 객체가 가진 속성 : __annotations__, __call__, __class__, __closure__, __code__, __default__, __delattr__, __dict__, __dir__, __doc__, __eq__, __format__, __ge__, __get__, __getattribute__, __globals__, __gt__, __hash__, __init__, __kwdefaults__, __le__, __lt__, __module__, __name__, __ne__, __new__, __qualname__, __reduce__, __reduce_ex__, __repr__, __setattr__, __sizeof__, __str__, __subclasshook__

__dict__ :  객체에 할당된 사용자 속성 보관. 기본적인 에너테이션 형태. 

프레임워크(장고,,,)는 어떤 속성을 함수에 할당하는 기능 사용

사용자 정의 객체에 없는 함수의 고유한 속성:
  __annotations__ : 자료형-dict, 매개변수 및 반환값에 대한 주석
  __call__ : 자료형-메서드-래퍼, 콜러블 객체 프로토콜에 따른 () 연산자 구현
  __closure__ : 자료형-튜플, 자유 변수 등 함수 클로저
  __code__ : 자료형-코드, 바이트코드로 컴파일된 함수 메타데이터 및 함수 본체
  __defaults__ : 자료형-튜플, 형식 매개변수의 기본값
  __get__ : 자료형-메서드-래퍼, 읽기 전용 디스크립터 프로토콜 구현
  __globals__ : 자료형-dict, 함수가 정의된 모듈의 전역 변수
  __kwdefaults__ : 자료형-dict, 키워드 전용 형식 매개변수의 기본값
  __name__ : 자료형-str, 함수명
  __qualname__ : 자료형-str, 전체 함수 명칭 (ex, random.choice() )
  
## 5.7. 위치 매개변수 ~ 키워드 전용 매개변수
  
키워드 전용 인수 >> 매개변수 처리 메커니즘 향상됨. (파이썬3)
  
(예 5-11) tag() 함수 호출
  
함수 정의시 키워드 전용 인수 지정하려면 >> * 붙은 인수 뒤에 이름 지정
가변 개수의 위치 인수 지원하지 않으며 키워드 전용 인수 지원하려면 >> * 시그니처에 포함

def f(a, * , b) : return a,b
f(1, b=2) # return (1,2)

키워드 전용인수 : 기본값 지정하지 않아도 됨. 필수 인수로 만들 수 있음.
